```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data= pd.read_csv('Dataset .csv')
```

```python
data.head(30)
```

```python
data.isnull().count()
```

```python
def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9\s]', '', text)
```

```python
data['Restaurant Name'] = data['Restaurant Name'].apply(remove_special_characters)
data['City'] = data['City'].apply(remove_special_characters)
data['Address'] = data['Address'].apply(remove_special_characters)
data['Locality'] = data['Locality'].apply(remove_special_characters)
data['Locality Verbose'] = data['Locality Verbose'].apply(remove_special_characters)
```

```python
data.head(30)
```

# Determine the percentage of restaurants that offer online delivery.

```python
data['Has Online delivery']
```

```python
data['Has Online delivery']=data['Has Online delivery'].map({'Yes': 1, 'No': 0})
```

```python
data['Has Online delivery'].value_counts()
```

```python
data['Has Online delivery'].sum()
```

```python
online_delivery = data['Has Online delivery'].sum()
```

```python
total_restaurant = len(data['Restaurant Name'])
```

```python
percentage= (online_delivery / total_restaurant)*100
```

```python
print("The percentage of restaurants that offer online delivery.")
print(percentage)
```
