```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data= pd.read_csv('Dataset .csv')
```

```python
data.head(30)
```

```python
data.isnull().count()
```

```python
def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9\s]', '', text)
```

```python
data['Restaurant Name'] = data['Restaurant Name'].apply(remove_special_characters)
data['City'] = data['City'].apply(remove_special_characters)
data['Address'] = data['Address'].apply(remove_special_characters)
data['Locality'] = data['Locality'].apply(remove_special_characters)
data['Locality Verbose'] = data['Locality Verbose'].apply(remove_special_characters)
```

```python
data.head(30)
```

# Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
data['Price range']
```

```python
range=data['Price range'].value_counts()
```

```python
plt.bar(range.index, range.values, color='green', width = 0.5)
plt.xlabel('Restaurant')
plt.ylabel('Price range')
plt.show()
```
