```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data= pd.read_csv('Dataset .csv')
```

```python
data.head(30)
```

```python
data.isnull().count()
```

```python
def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9\s]', '', text)
```

```python
data['Restaurant Name'] = data['Restaurant Name'].apply(remove_special_characters)
data['City'] = data['City'].apply(remove_special_characters)
data['Address'] = data['Address'].apply(remove_special_characters)
data['Locality'] = data['Locality'].apply(remove_special_characters)
data['Locality Verbose'] = data['Locality Verbose'].apply(remove_special_characters)
```

```python
data.head(30)
```

# Analyze the distribution of aggregate ratings and determine the most common rating range.

```python
data = data.dropna(subset=['Aggregate rating'])
```

```python
data['Aggregate rating'] = pd.to_numeric(data['Aggregate rating'], errors='coerce')
```

```python
data = data.dropna(subset=['Aggregate rating'])
```

```python
print(data['Aggregate rating'].describe())
```

```python
print(data['Aggregate rating'].unique())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Aggregate rating'], bins=10, kde=True)
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
rating_counts = data['Aggregate rating'].value_counts().sort_index()
```

```python
most_common_rating_range = rating_counts.idxmax()
```

```python
print(f"The most common rating is: {most_common_rating_range}")
```
